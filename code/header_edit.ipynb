```python
def make_filelist(path):
    import os

    file=[]
    for root, directories, files in os.walk(path, topdown=False):  
        for name in files:
            extens = os.path.splitext(name)[-1]
            if extens =='.fits':
                file.append(os.path.join(root,name))
        
    return file
```
