```python
###image extracting
###author : Yunyi Choi
###last edit : 20210720

def show_image(image_file):
    import numpy as np
    import matplotlib.pyplot as plt
    import astropy.io.fits as fits
    
    image_data = fits.open(image_file)
    data=image_data[0].data
    header=image_data[0].header

    image=np.array(data)


    max_value = np.percentile(image,99.8)
    min_value = np.percentile(image,15)

    plt.figure(figsize=(13,13))
    plt.imshow(image, cmap='gray', vmax = max_value, vmin = min_value, origin = 'lower')
    #plt.xlim(1000,1750)
    #plt.ylim(300,1000)
    #plt.show()
    plt.colorbar()

    print('Min:', np.min(image))
    print('Max:', np.max(image))
    print('Mean:', np.mean(image))
    #print('Stdev:', np.std(image)
```

```python
show_image('/data3/yunyi/SOAO/stacked_M101_20210604_r_mean.fits')
```
