```python
###magnitude from FITS
###author : Yunyi Choi
###last edit : 20210713


import numpy as np
import matplotlib.pyplot as plt
import astropy.io.fits as fits

specdata = fits.open('/data3/yunyi/SOAO/select/NGC5427/Calib-SOAO-NGC5427-20200318-202701-R-60.fits')
#specdata = fits.open('/data3/yunyi/SOAO/stacked_NGC3147_20201026_r.fits')
#specdata.info()

spectrum = specdata[0].data
header = specdata[0].header

flux = []
xpos = []

i=1153
for j in range(len(spectrum[i])):
    xpos.append(j)
    flux.append(spectrum[i][j])
    
    

skyval=np.mean(flux[:500])
print(skyval)


num=flux.index(max(flux))
max_f=max(flux)


flux = np.array(flux)
xpos = np.array(xpos)


plt.figure(figsize=(20,15))
plt.plot(xpos,flux,c='black')
plt.title('magnitude of BC56', fontsize=20)
plt.xlabel('x position', fontsize=15)
plt.ylabel('Flux', fontsize=15)
plt.axvspan(num-70,num+70,facecolor='r', alpha=0.5)
plt.text(num, max_f, max_f,fontsize=15, color='b')
```
