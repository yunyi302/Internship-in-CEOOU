```python
import matplotlib.pyplot as plt

depth = [20.6, 20.7, 19.9, 20.2, 19.9, 18.7, 19.8, 20.6]
exp_depth = [20.3, 20.5, 20.1, 20.0, 19.8, 18.6, 20.2, 20.6]
x=range(8)

dif_depth = []
mid_depth =[]
for i in range(8):
    dif_depth.append(abs(depth[i]-exp_depth[i]))
    mid = (depth[i]+exp_depth[i])/2
    mid_depth.append(mid)
    
#plt.figure(figsize=(20, 15))
plt.plot(x, depth, label='depth')
plt.plot(x, exp_depth, label='exp_depth')
for i in range(8):
    plt.text(x[i], mid_depth[i], round(dif_depth[i], 3))
plt.xlabel('image name', fontsize=14)
plt.ylabel('depth', fontsize=14)
plt.legend(fontsize=14)

plt.show()
```
