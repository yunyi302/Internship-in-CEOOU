```python
###magnitude from FITS
###author : Yunyi Choi
###last edit : 20210721


import numpy as np
import matplotlib.pyplot as plt
import astropy.io.fits as fits

specdata = fits.open('/data3/yunyi/SOAO/stacked_NGC3147_20201026_r_mean.fits')
#specdata.info()

spectrum = specdata[0].data
header = specdata[0].header


flux = []
xpos = []

i=833
for j in range(len(spectrum[i])):
    xpos.append(j)
    flux.append(spectrum[i][j])
    
    
    
num=flux.index(max(flux))
max_f=max(flux)
half_f=max_f/2


flux = np.array(flux)
xpos = np.array(xpos)

print(flux)
print(xpos)
print(half_f)
```

```python
half_x=[]


for i in range(len(flux)):
    if abs(flux[i]-half_f)<8000:
        print(flux[i])
        half_x.append(i)

print(half_x)

fwhm=half_x[1]-half_x[0]
ps=0.4454
fwhm_arcs = fwhm*ps
print(fwhm_arcs)
        
        
```

```python


plt.figure(figsize=(15,15))
plt.plot(xpos,flux,c='black')
plt.title('magnitude of BC56', fontsize=20)
plt.xlabel('x position', fontsize=15)
plt.ylabel('Flux', fontsize=15)
plt.axvspan(num-70,num+70,facecolor='r', alpha=0.5)
plt.text(num, max_f, max_f,fontsize=15, color='b')
```
